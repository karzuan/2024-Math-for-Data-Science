```python
#        
def derivative_x(f, x, step_size):
    m = (f(x + step_size) - f(x)) / ((x + step_size) - x)
    return m
def my_function (x):
    return x ** 2

slope_at_2 = derivative_x(my_function, 2, .00001) #   -        
print(slope_at_2) # prints 4.000010000000827
```

```python
#    
# X and step size
, s = symbols('x s')
# declare function
f= x**2
# slope between two points with step s
slope_f = (f.subs(x, x+s ) - f) / ((x+s) - x)
# substitute 2 for x
slope_2 = slope_f.subs(x, 2)
# calculate slope at x = 2
# by forever approaching a step size of o
result = limit(slope_2, s, 0)
print(result) # 4
```

### Gradient Descent with One Variable

```python
import random

def f(x):
    return (x - 3) ** 2 + 4

def dx_f(x):
    return 2*(x - 3)

L = 0.001 # The Learning rate
epochs = 100000 # The number of iterations to perform gradient descent
x = random.randint(-15,15) # start at a random x

for i in range(epochs):
    d_x = dx_f(x) # get sLope
    x= x - L * d_x # update x by subtracting the (Learning rate) * (slope)

print(x, f(x)) # prints 2.999999999999889 4.0
```

### Integrals

```python
# Reimann Sum method
def approximate_integral(a, b, n, f):
    delta_x = (b-a) / n
    total_sum = 0
    
    for i in range (1, n + 1):
        midpoint = 0.5 * (2 * a + delta_x * (2 * i - 1))
        total_sum += f(midpoint)    
    return total_sum * delta_x

def my_function(x):
    return x ** 2

area = approximate_integral(a=0, b=2, n=10000, f=my_function)

print(area) # prints 2.66666665999999
```

5. Assuming no payments are made, how much would a $1,000 loan be worth at 5% interest compounded monthly after 3 years?

```python
# approach when we add interest monthly w/o ajustment to the year. I.e. yearly is 60%
total = 1000
interest_pers = 0.05
months = 3*12

def debt(money, fee, time):
    for i in range(time):
        money += money*fee
    return money

print(debt(total, interest_pers, months))
```

```python
# approach when we add money montly, but we adjust to 5% yearly
total = 1000
interest_pers = 0.05
months = 3*12

def debt(money, fee, time):
    for i in range(time):
        money += (money*fee)/12
    return money

print(debt(total, interest_pers, months))
```
